```python
import random # the random module is used for generating random values
import string # the string module contains ASCII letters and digits
def main():
    passwords = read_passwords_from_file() #this jumps to this function read_passwords_from_file

    while True: # this is an infinite loop, the menu will be displayed repeatedly until the user chooses to exit the program.
        print("\n===================================") # the border of the menu
        print("Password Generator") # title of menu
        print("===================================") # separator of the menu
        print("1. Alphanumeric Password") # option 1 if the user wants to generate an alphamumeric password
        print("2. Alphanumeric with Symbols Password") # option 2 if the user wants to generate an alphamumeric with symbols password
        print("3. Show Stored Passwords") # option 3 if the user wants to see the generated passwords
        print("4. Delete Stored Password") # option 4 if the user wants to delete a saved password
        print("5. Exit") # option 5 if the user wants to exit
        print("===================================")

        choice = input("Enter choice (1-5): ") # the user enters a number from 1 to 5 depending on what they want to do

        if choice == '1': # if the user chooses option 1
            while True:
                length = input("Enter the length of the alphanumeric password: ") # this prompts the user to enter their desired length of the password
                if length.isdigit(): # if the user enters a valid number from 1 to 5
                    length = int(length) # this converts the user's input to an integer
                    break # this exits the inner loop because a vailid number is entered
                else:
                    print("Please enter a valid number for the password length.") # otherwise prompt the user to enter a valid number for choice again
            characters = string.ascii_letters + string.digits # this concatenates all English letters and numbers in text format '0123456789'
            generated_password = generate_password(length, characters) # this jumps to the generate_password(length, characters) function with length, characters as the arguments
            strength = strong_pass(generated_password, choice) # this jumps to the strong_pass(generated_password, choice) function with generated_password, choice as the arguments
            if strength == 'Weak': # if the strength of the password is weak, repeat the same process again go generate and test a new password
              characters = string.ascii_letters + string.digits
              generated_password = generate_password(length, characters)
              strength = strong_pass(generated_password, choice)
            else: # if the strength of the password is not weak
              name = input("Enter a name to store the password: ") # this prompts the user to enter their desired name to store the password
              passwords[name] = generated_password # this assigns generated_password to key name in dictionary passwords
              print(f"Generated Alphanumeric Password stored as '{name}': {generated_password} with strength: {strength}")
        elif choice == '2': # if the user chooses option 2
            while True: # ?
                length = input("Enter the length of the alphanumeric password with symbols: ") # this prompts the user to enter their desired length of the password
                if length.isdigit(): # if the user enters a valid number from 1 to 5
                    length = int(length) # this converts the user's input to an integer
                    break
                else:
                    print("Please enter a valid number for the password length.") # otherwise prompt the user to enter a valid number for choice again
            characters = string.ascii_letters + string.digits + string.punctuation
            generated_password = generate_password(length, characters)
            strength = strong_pass(generated_password, choice)
            if strength == 'Weak':
              characters = string.ascii_letters + string.digits + string.punctuation
              generated_password = generate_password(length, characters)
              strength = strong_pass(generated_password, choice)
            else:
              name = input("Enter a name to store the password: ")
              passwords[name] = generated_password
              print(f"Generated Alphanumeric Password with Symbols stored as '{name}': {generated_password} with strength: {strength}")
        elif choice == '3': # if the user chooses option 3
            show_passwords_table(passwords) # this jumps to show_passwords_table(passwords) function with passwords as argument
        elif choice == '4': # if the user chooses option 4
            delete_password(passwords) # this jumps to delete_password(passwords) functions with passwords as the argument
        elif choice == '5': # if the user chooses option 5
            write_passwords_to_file(passwords) # this jump to write_passwords_to_file(passwords) with passwords as the argument
            print("Exiting Password Generator. Goodbye!") # this prompts this sentence so the user knows they are existing the password generator program
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

def generate_password(length, characters): # define this functions with length, characters as parameters, to generate password
    password = ''.join(random.choice(characters) for _ in range(length)) # loop for 'length' times, choose randomly from 'characters' and then join those together to form the password
    return password # return the password

def delete_password(passwords): # define this function to delete password
    if passwords: # if password has at least one item
        print("Stored Passwords:")
        for name in passwords: # access all name (Facebook, Canvas,...) in dict passwords
            print(name) # print all names so the user can see what website's pass they want to delete
        name = input("Enter the name of the password to delete: ") # enter the name of the website whose password the user wants to delete
        if name in passwords: # if the website's name is in dict passwords
            del passwords[name] # delete the key name
            print(f"Password '{name}' deleted.") # print this to inform the user that password is deleted
        else:
            print("Password name not found.") # print this if password is not found in dict passwords
    else:
        print("No passwords stored yet.") # print this if the user has not chosen any option (1, 2) to generate passwords. There is nothing to delete

def show_passwords_table(passwords): # define this function to show a table of password
    if passwords: # if password has at least one item
        print("\nStored Passwords:")
        print("Name\t\tPassword\t\tStrength")
        print("-" * 40)  # Line separator
        for name, password in passwords.items():
            tabs_password = "\t" * (2 - (len(password) // 8)) # this calculates the number of tabs needed to align the password column based on its length.
            print(f"{name}\t\t{password}{tabs_password}")
            print("-" * 40)  # Line separator after each entry
    else:
        print("No passwords stored yet.")

def read_passwords_from_file(): # this function open the file in which passwords are stored
    try:
        with open('password.txt', 'r') as file: # open file 'password.txt' in read mode
            lines = file.readlines() # this reads all lines from the file and stores them in the lines variable as a list
            passwords = {} # create an empty dictionary
            for line in lines: # this iterates over each line in the list of lines obtained from the file.
                data = line.strip().split(';') # each line is stripped of leading and trailing whitespace using strip(), and then it is split into a list using the semicolon (';') as a delimiter
                if len(data) >= 2:  # Check if there are enough elements in the split line
                    name = data[0] # this extracts the name
                    password = data[1] # this extracts the password
                    passwords[name] = password # this stores name as the key and password as the value in the password dictionary
            return passwords # this returns the passwords dictionary containing the stored names and passwords
    except FileNotFoundError: # if the 'password.txt' file is not found
        return {} # an empty dictionary is returned

def write_passwords_to_file(passwords): #this function writes generated password to the password.txt file
    with open('password.txt', 'w') as file: # this creates the password.txt file
        for name, password in passwords.items(): # this loop accesses name and password in the password dictionary
            file.write(f"{name};{password}\n") # this writes name and password to the file


def strong_pass(password,choice): # this function checks the strength of the password
    if choice == '1':
        alpha = not(password.isalpha()) # The password should contain at least one non-alphabetic character
        upper = not(password.isupper()) # The password should contain at least one lowercase letter
        lower = not(password.islower()) # The password should contain at least one uppercase letter
        num   = not(password.isdigit()) # The password should contain at least one non-digit character
        strong = alpha and upper and lower and num # the password is strong if all conditions above are met
        if strong == True:
            return('Strong')
        else:
            return('Weak')
    if choice == '2':
        alpha = not(password.isalpha())
        upper = not(password.isupper())
        lower = not(password.islower())
        num   = not(password.isdigit())
        alnum = not(password.isalnum()) # The password should contain at least one non-alphanumeric character
        strong = alpha and upper and lower and num and alnum
        if strong == True:
            return('Strong')
        else:
            return('Weak')

if __name__ == "__main__": # this checks whether the script is being run directly, not imported as a module into another script
    main() # if it passes the check, call the main() function
```
